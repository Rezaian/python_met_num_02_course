```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('fivethirtyeight') 


# Generate 200 points between 0 and 2*pi
x = np.linspace(0, 2*np.pi, 200)

# Calculate sin values for each point
y = np.sin(x)

# Plot the sine wave
plt.plot(x, y)
plt.title('Sine Wave')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight') 

# Data for plotting
years = list(range(2014, 2024))
temperatures = [0.13, 0.3, 0.35, 0.26, 0.16, 0.31, 0.34, 0.22, 0.24, 0.54]
plt.bar(years, temperatures, label="Global") 

temperature_nh = [0.15, 0.38, 0.0, 0.34, 0.19, 0.38, 0.51, 0.35, 0.35, 0.68]
plt.bar(years, temperature_nh, linestyle='--', label="North Hemisphere") 

plt.title('Global Average Temperature Anomalies (2010 - 2023)')
plt.xlabel('Year')
plt.ylabel('Temperature Anomaly (°C)')

plt.legend()
plt.tight_layout()

# Show the plot
plt.show()
```

```python
#apply index
# set width
# add ticks
#rotate ticks

import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight') 

# Data for plotting
years = list(range(2014, 2024))
index = np.arange(len(years)) ###
print(index)
width = 0.25 ###

temperatures = [0.13, 0.3, 0.35, 0.26, 0.16, 0.31, 0.34, 0.22, 0.24, 0.54]
plt.bar(index, temperatures, width = width, label="Global")  

temperature_nh = [0.15, 0.38, 0.43, 0.34, 0.19, 0.38, 0.51, 0.35, 0.35, 0.68]
plt.bar(index - width, temperature_nh, width = width,label="North Hemisphere") 

temperature_sh = [0.1, 0.2, 0.26, 0.17, 0.13, 0.23, 0.16, 0.09, 0.11, 0.38]
plt.bar(index + width, temperature_sh, width = width, label="South hemisphere")

plt.title('Global Average Temperature Anomalies (2010 - 2023)')
plt.xlabel('Year')
plt.ylabel('Temperature Anomaly (°C)')

plt.legend()
plt.tight_layout()

plt.xticks(ticks=index, labels=years) ###
plt.xticks(rotation=90) ###

# Show the plot
plt.show()
```
