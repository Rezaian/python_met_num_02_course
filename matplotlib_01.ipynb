```python
#add text title

import matplotlib.pyplot as plt

# Data for plotting
years = list(range(2014, 2024))
#source: https://ds.data.jma.go.jp/tcc/tcc/products/gwp/temp/list/year_wld.html
temperatures = [0.13, 0.3, 0.35, 0.26, 0.16, 0.31, 0.34, 0.22, 0.24, 0.54]
plt.plot(years, temperatures)

plt.title('Global Average Temperature Anomalies (2010 - 2023)') ###
plt.xlabel('Year') ### 
plt.ylabel('Temperature Anomaly (°C)') ###

# Show the plot
plt.show()
```

```python
#add more data 
#define linewidth

import matplotlib.pyplot as plt
# Data for plotting
years = list(range(2014, 2024))
temperatures = [0.13, 0.3, 0.35, 0.26, 0.16, 0.31, 0.34, 0.22, 0.24, 0.54]
###
plt.plot(years, temperatures, color='b', marker='.',  linewidth=3,label="Global") 

temperature_nh = [0.15, 0.38, 0.43, 0.34, 0.19, 0.38, 0.51, 0.35, 0.35, 0.68]
###
plt.plot(years, temperature_nh, color='k', linestyle='--', marker='o', label="North Hemisphere")

temperature_sh = [0.1, 0.2, 0.26, 0.17, 0.13, 0.23, 0.16, 0.09, 0.11, 0.38]###
###
plt.plot(years, temperature_sh, color='m', linestyle='-.', marker='*', label="South hemisphere")

plt.title('Global Average Temperature Anomalies (2010 - 2023)')
plt.xlabel('Year')
plt.ylabel('Temperature Anomaly (°C)')

plt.legend()
# Show the plot
plt.show()
```

```python
#add grid
#tight_layout

import matplotlib.pyplot as plt
# Data for plotting
years = list(range(2014, 2024))
temperatures = [0.13, 0.3, 0.35, 0.26, 0.16, 0.31, 0.34, 0.22, 0.24, 0.54]
plt.plot(years, temperatures, color='b', marker='.',  linewidth=2,label="Global") 

temperature_nh = [0.15, 0.38, 0.43, 0.34, 0.19, 0.38, 0.51, 0.35, 0.35, 0.68]
plt.plot(years, temperature_nh, color='k', linestyle='--', marker='o', label="North Hemisphere")

temperature_sh = [0.1, 0.2, 0.26, 0.17, 0.13, 0.23, 0.16, 0.09, 0.11, 0.38]
plt.plot(years, temperature_sh, color='m', linestyle='-.', marker='*', label="South hemisphere")

plt.title('Global Average Temperature Anomalies (2010 - 2023)')
plt.xlabel('Year')
plt.ylabel('Temperature Anomaly (°C)')

plt.legend()
plt.grid(True)###
plt.tight_layout()###
# Show the plot
plt.show()
```
